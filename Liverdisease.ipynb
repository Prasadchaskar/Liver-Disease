```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
liver_df = pd.read_csv('indian_liver_patient.csv')
liver_df.head()
```

```python
liver_df.columns
```

```python
liver_df.info()
```

```python
liver_df.rename(columns={'Dataset':'Label'},inplace=True)
liver_df.head()
```

```python
liver_df.isnull().sum()
```

```python
sns.countplot(x='Label',data=liver_df);
```

```python
liver_df["Albumin_and_Globulin_Ratio"] = liver_df.Albumin_and_Globulin_Ratio.fillna(liver_df['Albumin_and_Globulin_Ratio'].mean())
```

```python
liver_df.head()
```

```python
def func(df):
    if df=='Male':
        return 0
    else:
        return 1
liver_df.Gender = liver_df.Gender.apply(func)
```

```python
liver_df.head()
```

```python
X = liver_df.drop('Label',axis=1)
y = liver_df.Label
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier

models = {
    LogisticRegression(max_iter=1000):'Logistic Regression',
    SVC():"Support Vector Machine",
    RandomForestClassifier(n_estimators=100):'Random Forest',
    GradientBoostingClassifier():'xg'
}
for m in models.keys():
    m.fit(X_train,y_train)
for model,name in models.items():
     print(f"Accuracy Score for {name} is : ",model.score(X_test,y_test)*100,"%")
```
